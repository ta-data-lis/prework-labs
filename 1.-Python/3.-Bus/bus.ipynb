# Bus

This bus has a passenger entry and exit control system to monitor the number of occupants it carries and thus detect when there are too many.

At each stop, the entry and exit of passengers is represented by a tuple consisting of two integer numbers.
```
bus_stop = (in, out)
```
The succession of stops is represented by a list of these tuples.
```
stops = [(in1, out1), (in2, out2), (in3, out3), (in4, out4)]
```

## Tasks

````python
# Variables
stops = [(10, 0), (4, 1), (3, 5), (3, 4), (5, 1), (1, 5), (5, 8), (4, 6), (2, 3)]
````

#### 1. Calculate the number of stops.

````python
print("This bus has", len(stops), "stops.")
````

#### 2. Assign to a variable a list whose elements are the number of passengers at each stop (in-out).
Each item depends on the previous item in the list + in - out.

````python
passenger_count = []
stop_count = 0
stop_number = range(0,9)
stop_in = [x[0] for x in stops]
stop_out = [x[1] for x in stops]

for x in stop_number:
    stop_count += stop_in[x] - stop_out[x]
    passenger_count.append(stop_count)
    
print(passenger_count)
````

#### 3. Find the maximum occupation of the bus.

````python
passenger_max = max(passenger_count)

print("The maximum occupation of the bus is", passenger_max, "passengers.")
````

#### 4. Calculate the average occupation. And the standard deviation.

````python
passenger_avg = sum(passenger_count) / len(passenger_count)

print("The average occupation of the bus is", passenger_avg,"passengers.")

passenger_dev = []

for x in stop_number:
    passenger_dev.append((passenger_count[x] - passenger_avg)**2)
    
passenger_std = (sum(passenger_dev) / len(passenger_dev))**0.5

print("The standard deviation is", passenger_std, "passengers.")
````
